```python
basenames= []
for i in range(1,4):
    basenames.append("dataset/base_data"+str(i)+".sas7bdat")
```

```python
base_list = []
for i, file in enumerate(basenames):
    with SAS7BDAT(file) as file: 

        base_list.append(file.to_data_frame())
        print(f"{base_list[i].shape=}")
```

```python
len(base_list)
```

```python
cnt = 0
for df in base_list:
    if cnt == 0 :
        base_merge_df = df.copy()
        continue
#     base_merge_df = pd.merge(base_merge_df,df,on="T_ID",left_index=True, right_index=True)
    base_merge_df = base_merge_df.join(df)
    cnt +=1
```
